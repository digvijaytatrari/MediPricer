```python
# Code for checking is the crome driver working properly or not 
from selenium import webdriver   #using selenium for automation
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
```

```python
# find url for every combination of ciys and specialty

specialty_list = [
    'Dentist', 'Gynecologist', 'Pediatrician', 'Orthopedist', 'Dermatologist', 
    'Physiotherapist', 'Infertility Specialist', 'Ophthalmologist', 'Urologist', 
    'Psychiatrist', 'Cardiologist', 'Dietetian', 'Gastroenterologist', 'Neurologist', 
    'Neurosurgeon', 'Pulmonologist', 'Bariatric Surgeon', 'Rheumatologist', 'Chiropractor'
]

citys = ['Delhi', 'Mumbai', 'Bangalore']

urls = []

# Loop through every city for every specialty
for city in citys:
    for specialty in specialty_list:
        
        test = [specialty]
        
        # If specialty contains a space, replace it with "%20"
        if ' ' in specialty:
            specialty = specialty.replace(' ', '%20')

        if specialty == 'Dietetian':
            url = f'https://www.practo.com/search/doctors?results_type=doctor&q=%5B%7B%22word%22%3A%22dietetian%22%2C%22autocompleted%22%3Atrue%2C%22category%22%3A%22common_name%22%7D%5D&city={city}'
        else:  
            url = f'https://www.practo.com/search/doctors?results_type=doctor&q=%5B%7B%22word%22%3A%22{specialty}%22%2C%22autocompleted%22%3Atrue%2C%22category%22%3A%22subspeciality%22%7D%5D&city={city}'
        
        test.append(url)
        urls.append(test)
```

```python
print(len(urls))
```

```python
options = webdriver.ChromeOptions()
options.add_argument("Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36")

driver = webdriver.Chrome()

# Data to store doctor's info
data = []

for url in urls:

    driver.get(url[1])  

    # Set a delay to allow time for content to load
    SCROLL_PAUSE_TIME = 3

    # Get the initial scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight / 2);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Scroll down the remaining half
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # Scrapping Data

    doc_info = driver.find_elements(By.CLASS_NAME, 'info-section')
    

    
    # count to only load 100 data for each url
    count = 0
    
    for info in doc_info:
        try:
            link = info.find_element(By.TAG_NAME, 'a')

            # Get the current window handle
            original_window = driver.current_window_handle

            # Click the link (which may open a new tab)
            link.click()

            time.sleep(2)

            # Switch the WebDriver focus to the new tab
            driver.switch_to.window(driver.window_handles[-1])

            try:
                # Find the doctor's name element
                name = driver.find_element(By.XPATH,'//h1[@data-qa-id="doctor-name"]').text.strip()
            except NoSuchElementException:
                name = "Not found"

            try:
                # Find the degree type element
                degree_type = driver.find_element(By.XPATH,'//p[@data-qa-id="doctor-qualifications"]').text.strip()
            except NoSuchElementException:
                degree_type = "Not found"

            try:
                # Find the years of experience element
                year_of_experience = driver.find_element(By.XPATH, '//h2[contains(span, "Years Experience Overall")]').text.strip()
            except NoSuchElementException:
                year_of_experience = "Not found"

            try:
                # Find the specialty of the doctor element
                speciality_of_doctor = url[0]
            except NoSuchElementException:
                speciality_of_doctor = "Not found"

            try:
                # Find the location element
                location = driver.find_element(By.XPATH, '//h4[@data-qa-id="clinic-locality"]').text.strip()
            except NoSuchElementException:
                location = "Not found"

            try:
                # Find the dp score element
                dp_score = driver.find_element(By.XPATH, '//span[@class="u-green-text u-bold u-large-font"]').text.strip()
            except NoSuchElementException:
                dp_score = "Not found"

            try:
                # Find the npv element
                npv = driver.find_element(By.XPATH, '//span[@class="u-smallest-font u-grey_3-text"]').text.strip()
            except NoSuchElementException:
                npv = "Not found"

            try:
                # Find the consultation fee element
                consult_fee = driver.find_element(By.XPATH, '//div[@class="u-f-right u-large-font u-bold u-valign--middle u-lheight-normal"]').text.strip()
            except NoSuchElementException:
                consult_fee = "Not found"

            time.sleep(1)
            
            # Dictonary to store data
            doc_data = {}
            
            #Store inside dictonary
            doc_data['name'] = name
            doc_data['degree_type'] = degree_type
            doc_data['year_of_experience'] = year_of_experience
            doc_data['speciality_of_doctor'] = speciality_of_doctor
            doc_data['location'] = location
            doc_data['dp_score'] = dp_score
            doc_data['npv'] = npv
            doc_data['consult_fee'] = consult_fee

            # Add doctor's data to list 
            data.append(doc_data)
            
            # Increment count
#             count += 1
            
            # Break the loop if count exceeds or equals 100
#             if count >= 100:
#                 break  
                
            # close the new tab
            driver.close()
            
#             break
            
            # Switch back to the original tab (if needed)
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            print(f'not found')


      
```

```python
      
            
            
            
# Converting Scraped data into Excel file
            
import pandas as pd
import os

# Extracting relevant information
names = [entry['name'] for entry in data]
degrees = [entry['degree_type'] for entry in data]
years_of_experience = [entry['year_of_experience'] for entry in data]
specialities = [entry['speciality_of_doctor'] for entry in data]
locations = [entry['location'].split(",")[0] for entry in data]
city = [entry['location'].split(",")[-1] for entry in data]
dp_scores = [entry['dp_score'] for entry in data]
npvs = [entry['npv'] for entry in data]
consult_fees = [entry['consult_fee'] for entry in data]

# Creating DataFrame
df = pd.DataFrame({
    'Name': names,
    'Degree Type': degrees,
    'Years of Experience': years_of_experience,
    'Speciality': specialities,
    'Location': locations,
    'City': city,
    'DP Score': dp_scores,
    'NPV': npvs,
    'Consult Fee': consult_fees
})

# Save DataFrame to CSV
df.to_csv('doctors_info.csv', index=False)

directory = 'Downloads'

# Save DataFrame to CSV inside directory
csv_file_path = os.path.join(directory, 'full_docters_info.csv')
df.to_csv(csv_file_path, index=False)
```
