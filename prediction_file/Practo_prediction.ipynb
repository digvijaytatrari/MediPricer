### Scaling

```python
from sklearn.model_selection import train_test_split


train_val_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=42)
```

```python
categorical = df.select_dtypes('object').columns.tolist()
numerical = df.select_dtypes('number').columns.tolist()

numerical.remove('Consult Fee')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit the scaler on the training data and transform the numerical features
train_df[numerical] = scaler.fit_transform(train_df[numerical])

# Transform the numerical features in the validation and testing sets
test_df[numerical] = scaler.transform(test_df[numerical])
val_df[numerical] = scaler.transform(val_df[numerical])
```

```python
# No need to use Encoding because No categorical columns exists

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(df[categorical])

encoded_cols = list(encoder.get_feature_names_out(categorical))
```

```python
encoded_cols
```

```python
inputs = df.columns.tolist()[:-1]
target = 'Consult Fee' 

train_inputs = train_df[inputs]
train_targets = train_df[target]

test_inputs = test_df[inputs]
test_targets = test_df[target]

val_inputs = val_df[inputs]
val_targets = val_df[target]


train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical])
```

### 2. Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Creating and fitting the Decision Tree Regressor model
model = DecisionTreeRegressor(
    max_depth=5,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features=None,
    random_state=42
)
model.fit(train_inputs[numerical + encoded_cols], train_targets)

# Predicting on the training and validation sets
train_pred = model.predict(train_inputs[numerical + encoded_cols])
val_pred = model.predict(val_inputs[numerical + encoded_cols])

# Calculate evaluation metrics
train_mse = mean_squared_error(train_targets, train_pred)
val_mse = mean_squared_error(val_targets, val_pred)
train_r2 = r2_score(train_targets, train_pred)
val_r2 = r2_score(val_targets, val_pred)
train_mae = mean_absolute_error(train_targets, train_pred)
val_mae = mean_absolute_error(val_targets, val_pred)

# Displaying all metrics
print("Training MSE:", train_mse)
print("Validation MSE:", val_mse)
print()
print("Training R-squared:", train_r2)
print("Validation R-squared:", val_r2)
print()
print("Training MAE:", train_mae)
print("Validation MAE:", val_mae)
```

### 3. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Create a RandomForestRegressor model
model = RandomForestRegressor(random_state=42)

# Define the hyperparameters grid to search
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'max_features': ['sqrt', 'log2']  # Number of features to consider when looking for the best split
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search.fit(train_inputs[numerical + encoded_cols], train_targets)

# Get the best model and predict on the test data
best_model = grid_search.best_estimator_
y_pred = best_model.predict(test_inputs[numerical + encoded_cols])

# Print the best hyperparameters found by grid search
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
# Calculate evaluation metrics
train_mse = mean_squared_error(train_targets, train_pred)
val_mse = mean_squared_error(val_targets, val_pred)
train_r2 = r2_score(train_targets, train_pred)
val_r2 = r2_score(val_targets, val_pred)
train_mae = mean_absolute_error(train_targets, train_pred)
val_mae = mean_absolute_error(val_targets, val_pred)

# Displaying all metrics
print("Training MSE:", train_mse)
print("Validation MSE:", val_mse)
print()
print("Training R-squared:", train_r2)
print("Validation R-squared:", val_r2)
print()
print("Training MAE:", train_mae)
print("Validation MAE:", val_mae)
```
